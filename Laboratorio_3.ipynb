```python
import pandas as pd
df = pd.read_csv("/content/sample_data/1980sClassics.csv")
df
```

```python
# Convierte el tiempo "minutos:segundos" a segundos
def convertir_a_segundos(duracion):
    minutos, segundos = map(int, duracion.split(":"))
    return minutos * 60 + segundos

df_2 = df.copy()
df_2["Duration"] = df_2["Duration"].apply(convertir_a_segundos)

print("Canciones con duracion mayor a 4")
canciones = df_2[df_2["Duration"] > 240]
canciones
```

```python
print("Canciones del artista Cliff Richard en el año 1980")
canciones = df[(df["Artist"] == "Cliff Richard") & (df["Year"] == 1980)]
canciones
```

```python
print("Canciones del artista Michael Jackson en el año 1989")
canciones = df[(df["Artist"] == "Michael Jackson") & (df["Year"] == 1989)]
canciones
```

```python
print("Canciones con popularidad mayor a 90 del año 1981")
canciones = df[(df["Popularity"] > 80) & (df["Year"] == 1981)]
canciones
```

```python
print("Canciones con positividad musical menor a 0.352")
canciones = df[(df["Valence"] < 0.352)]
canciones
```

```python
print("Canciones de Billy Joel con tonalidad G(sol)")
canciones = df[(df["Artist"] == "Billy Joel") & (df["Key"] == 7)]
canciones
```

```python
print("Ingrese el nombre del artista que desea buscar")
input = input()
canciones = df[(df["Artist"] == input)]
canciones
```
